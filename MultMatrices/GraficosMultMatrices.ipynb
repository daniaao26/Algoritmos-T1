```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
#Gráfico de matrices de la misma dimensión con un rango de números entre -1000 y 1000

# Iterativo Tradicional
y2 = np.array([0, 0, 0.01, 11.214, 14.008, 101.757])

# Iterativo Optimizado
y3 = np.array([0, 0, 0.011, 11.535, 12.514, 94.689])

# Strassen
y4 = np.array([0, 0.008, 2.675, 494.432, 500.703, 3499.93])

# Tamaños de las matrices (se usa solo la primera dimensión)
x = np.array([4, 10, 100, 1000, 1024, 2000])

plt.plot(x, y2, label='Iterativo Tradicional')
plt.plot(x, y3, label='Iterativo Optimizado')
plt.plot(x, y4, label='Strassen')

plt.ylabel('Tiempo (segundos)')
plt.xlabel('Tamaño de la matriz (NxN)')
plt.title('Comparación de algoritmos de multiplicación de matrices')

plt.legend()
plt.grid(True)

plt.show()
```

```python
#Gráfico de matrices de distinta dimensión con un rango de números entre -1000 y 1000

y_tradicional = np.array([0, 0, 0.017, 4.478, 16.699, 89.167])
y_optimizado = np.array([0, 0, 0.008, 2.463, 11.775, 85.233])

# Tamaños de las matrices, es el producto de las dimensiones
x = np.array([3*5*3, 20*10*20, 80*120*80, 512*900*512, 1000*1024*1000, 1800*2048*1800])

plt.plot(x, y_tradicional, label='Iterativo Tradicional')
plt.plot(x, y_optimizado, label='Iterativo Optimizado')

plt.ylabel('Tiempo (segundos)')
plt.xlabel('Tamaño de la multiplicación (AxBxC)')
plt.title('Comparación de algoritmos de multiplicación de matrices')

plt.xscale('log')#Mejor visualización de los datos ya que el tamaño de las multiplicaciones puede ser muy grande

plt.legend()
plt.grid(True)

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
#Gráfico de números aleatorios con límite de numeros entre -1000 y 1000

# Bubblesort
#y1 = np.array([0.003, 0.366, 44.464, 26767.8])
#x1 = np.array([10**3, 10**4, 10**5, 10**6])

# Mergesort
y2 = np.array([0.001, 0.004, 0.048, 0.89])
x2 = np.array([10**3, 10**4, 10**5, 10**6])

# Quicksort
y3 = np.array([0, 0.004, 0.063, 3.802])
x3 = np.array([10**3, 10**4, 10**5, 10**6])

# Sort
y4 = np.array([0, 0.001, 0.023, 0.231])
x4 = np.array([10**3, 10**4, 10**5, 10**6])

#plt.plot(x1, y1, label='BubbleSort')
plt.plot(x2, y2, label='MergeSort')
plt.plot(x3, y3, label='QuickSort')
plt.plot(x4, y4, label='Sort')

plt.ylabel('Tiempo (segundos)')
plt.xlabel('Tamaño del arreglo')
plt.title('Comparación de algoritmos de ordenamiento')
plt.legend()
plt.grid(True)

plt.show()
```
